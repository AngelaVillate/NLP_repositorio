#### Comenzaré por el análisis del texto de "Don Quijote" Extracto del primer capítulo 

Mi propósito es el de leer un extrato del primer capítulo e identificar el universo léxico y su frecuencia

```python
import re
```

```python
with open("xixote.txt", encoding='utf-8') as archivo: 
    data = archivo.read()
```

```python
print (data)
```

```python
data = data.lower()
print(data)
```

#### Conservar solo palabras

```python
data = re.sub(r"[\W\d_]+", " ", data)
print(data)
```

#### tokenizamos

```python
data = data.split()
data
```

#### Quitamos palabras vacías

```python
from nltk.corpus import stopwords
stopwords_sp= stopwords.words('spanish')
```

```python
data = [palabra for palabra in data if palabra not in stopwords_sp]
data
```

```python
from collections import Counter
lista=Counter(data).most_common(5000)
lista
```

#### En este paso convertimos a dataframe para pasar a excel y poder tener manejo para visualización

```python
import pandas as pd
```

```python
df= pd.DataFrame(lista, columns=['palabra','cantidad'])
df
```

```python
df.to_excel('PalabrasQuijote.xls')
```

### Ocurrencias por oración

```python
import es_core_news_sm
```

#### Sin contar palabras vacías

```python
nlp = es_core_news_sm.load()
doc = nlp(" ".join(data))

data_pos = [w.pos_ for w in doc] # [(w.text, w.pos_) for w in doc]

TipoPalabra=Counter(data_pos).most_common()
TipoPalabra
```

```python
df2= pd.DataFrame(TipoPalabra, columns=['Tipopalabra','cantidad'])
df2
```

```python
df2.to_excel('TipoPalabrasQjt.xls')
```

#### Con palabras vacías

```python
with open("xixote.txt", encoding='utf-8') as archivo: 
    data = archivo.read()
    
data = data.lower()
data = re.sub(r"[\W\d_]+", " ", data)

doc = nlp(data)

data_pos = [w.pos_ for w in doc] # [(w.text, w.pos_) for w in doc]

TotalTipo=Counter(data_pos).most_common()
TotalTipo
```

```python
df3= pd.DataFrame(TotalTipo, columns=['palabra','cantidad'])
df3
```

```python
df3.to_excel('TotalTipoPalabrasQjt.xls')
```
