```python
contador=0
for filas in dialogos["Locución"] : 
    #campo=dialogos["mensaje"][contador]
    nuevo_string= filas.lower()
    nuevo_string= re.sub(r"[\W\d_]+", " ", nuevo_string)
    token=nuevo_string.split()
    cadena=""
    for filas2 in token :
        if filas2 in stopwords_sp:
            cadena+=""
            continue  
        else : 
            cadena +=" "+filas2 
    dialogos["mensaje"][contador] = cadena
    cadena=""
    contador+=1
    if(contador==959):
        break
    continue
```

```python
# Puntos teóricos:

#  ¿En qué casos es buena idea tomar en la cuenta la frecuencia de las palabras para la bolsa de palabras?
Es buena idea para establecer contextos de uso lingüístico. Es decir, la frecuencia de palabras nos ayuda conocer (así sea de forma genérica)
la temática de la que trata el corpus que estamos analizando. En este punto es muy importante identificar si estamos frente a léxico técnico, 
de especialidad o de lenguaje común. Si estamos frente a un texto que pertenezca a una especialidad, el conteo de frecuencia de palabras
nos servirá para saber la naturaleza del corpus.Así, por ejemplo, si la palabra "paciente" se repite muchas veces 
estaremos frente a un contexto de uso de medicina
```

```python
#¿Cuándo es una mejor idea usar una bolsa de n-gramas en vez de una bolsa de palabras?
Yo preferiría una bolsa de ngramas para analizar si estamos frente a lenguas de tipo aglutinante o no, como por ejemplo
el alemán. 
```
