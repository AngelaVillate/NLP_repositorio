```python
import pandas as pd
data = pd.read_excel("Taller1.xlsx")
data.head()
```

```python
data_colum = pd.DataFrame(data[['¿Por qué estás estudiando análitica estratégica de datos?']])
for row in data_colum.itertuples(index=True, name='Pandas') :
    texto = "texto" + row[1]
```

```python
arreglo = texto.split()
arreglo
```

```python
import collections
words_count = collections.Counter(arreglo)
```

```python
words_count
```

```python
dataframe_org=pd.Series(words_count).to_frame('ColumnName')
dataframe_org
```

```python
dataframe_ord=dataframe_org.sort_values(by=['ColumnName'], ascending=False)
dataframe_ord.head(30)
```
